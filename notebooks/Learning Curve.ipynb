# Import Packages

```python
from MLTrainer import LearningCurve
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

# Generate Classification Dataset

```python
X, y = make_classification(random_state=42, n_samples=10000, n_features=5, n_classes=3,
                          n_informative=2, n_clusters_per_class=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Learning Curve

Parameters of each method are explained clearly in the source code.

.cmopute_scores method outputs mean cross validation score and test score for each sample size for a particular model

.plot method plots a learning curve using the Pandas DataFrame ouput from previous method

```python
curve = LearningCurve(batch_size=200, cv=5, seed=42, shuffle=False)
```

```python
model = RandomForestClassifier(random_state=42)
df = curve.compute_scores(model, X_train, y_train, X_test, y_test, scoring="accuracy")
```
